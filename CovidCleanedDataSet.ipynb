```python
import pandas as pd
import numpy as np
pol = pd.read_csv('Covid_stats_Jan2025.csv', encoding='ISO-8859-1')

print(pol.head())

```

```python
df = pol.drop(columns=['New Cases/1M pop','New Deaths/1M pop', 'NewCases','NewDeaths', 'NewRecovered'])
```

```python
df['TotalRecovered'] = pd.to_numeric(df['TotalRecovered'].astype(str).replace(',', '', regex = True), errors = 'coerce')
df['TotalCases'] = pd.to_numeric(df['TotalCases'].astype(str).replace(',', '', regex = True), errors = 'coerce')

valid_recovery = df.dropna(subset=['TotalCases', 'TotalRecovered'])
mean = (valid_recovery['TotalRecovered']/valid_recovery['TotalCases']).mean()
```

```python
df.head()
df['mean_recovery_rate'] = (valid_recovery['TotalRecovered']/valid_recovery['TotalCases'])*100
```

```python
print("before ", df['TotalRecovered'].isna().sum())
df.head()
```

```python
#filling of total recovered with mean precentage repective to population
df.loc[df['TotalRecovered'].isna(), 'TotalRecovered'] = df.loc[df['TotalRecovered'].isna(), 'TotalCases'] * mean/100
df.head()
```

```python

# filling null values in TotalDeath to 0
df.loc['TotalDeaths'] = pd.to_numeric(df['TotalDeaths'].astype(str).replace(',', '', regex = True), errors = 'coerce')
df['TotalDeaths'] = df['TotalDeaths'].fillna(0)
df.head()
```

```python
#filling of active cases
df['ActiveCases'] = pd.to_numeric(df['ActiveCases'].astype(str).replace(',', '', regex = True), errors = 'coerce')
df['TotalDeaths'] = pd.to_numeric(df['TotalDeaths'].astype(str).replace(',', '', regex = True), errors = 'coerce')
df['TotalRecovered'] = pd.to_numeric(df['TotalRecovered'].astype(str).replace(',', '', regex = True), errors = 'coerce')
mask = df['ActiveCases'].isna()
df.loc[mask, 'ActiveCases']  = df['TotalCases'] - df['TotalRecovered'] - df['TotalDeaths'] 
df.head()
```

```python
#filling total tests
df['TotalTests'] = pd.to_numeric(df['TotalTests'].astype(str).replace(',', '', regex = True), errors = 'coerce')

df['ratioOfInfection'] = df['TotalTests'] / df['TotalCases']
```

```python
meanInfection = df['ratioOfInfection'].mean()
```

```python
df['TotalTests'] = df['TotalTests'].fillna(df['TotalCases'] * meanInfection )
```

```python
#dropping irrelevant columns
df = df.drop(columns=[ 'TotÿCases/1M pop', 'Deaths/1M pop', '1 Caseevery X ppl', '1 Deathevery X ppl', 'ratioOfInfection' ])
```

```python
#filling population
df['Population'] = pd.to_numeric(df['Population'].astype(str).replace(',', '', regex = True), errors = 'coerce')
df['ratioForPopulation'] = df['Population'] / df['TotalCases'] 
```

```python
meanRatioForPopulation = df['ratioForPopulation'].mean()
```

```python
df.loc[df['Population'].isna(), 'Population'] = df.loc[df['Population'].isna(), 'Population'].fillna(df['TotalCases'] * meanRatioForPopulation)
df.head()
```

```python
df = df.drop(columns=[ 'ratioForPopulation', 'Serious,Critical', 'Tests/\n1M pop', '1 Testevery X ppl', 'ratioForPopulation', 'mean_recovery_rate', 'Active Cases/1M pop'])
df.isnull().sum()
```

```python
df.to_csv('cleanedCovidData.csv', index=False)
```
